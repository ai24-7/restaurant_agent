```python
import random
def random_price_value(price_val):
    # Handle missing values
    if pd.isna(price_val):
        return None
    # Convert to string and strip spaces
    price_str = str(price_val).strip()
    # Check that there is at least one dollar sign
    if '$' not in price_str:
        return None
    # Count the number of dollar signs
    count = price_str.count('$')
    if count == 1:
        return random.randint(1, 9)
    elif count == 2:
        return random.randint(11, 99)
    elif count == 3:
        return random.randint(101, 999)
    elif count == 4:
        return random.randint(1001, 9999)
    else:
        return None

# Apply the function to create a new numeric price column
df['price_numeric'] = df['price'].apply(random_price_value)
```
